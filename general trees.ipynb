### Node and tree construction

```python
from collections import deque
```

```python
class TreeNode():
    def __init__(self, x:str, k:int)->None:
        self.val = x
        self.arity = k
        self.child = [None]*k
```

```python
class SearchTree():
    def __init__(self, root: TreeNode) -> None:
        self.root = root
    
    def visit(self, depth, node:TreeNode):
        print(depth*'--', node.val)
        
    def BFT(self):
        if self.root == None:
            return
        q = deque([self.root])
        while q:
            currnode = q.popleft()       # 왼쪽부터 pop(방문)
            self.visit(currnode)
            for childnode in currnode.child:
                if childnode:
                    q.append(childnode)  # 왼쪽부터 달린 child node를 append
                    
    def __DFT_preorder_help(self, currnode:TreeNode, level = 0):
        if currnode == None:
            return
        self.visit(level, currnode)             # currnode가 none 아니면 visit
        for childnode in currnode.child:
            self.__DFT_preorder_help(childnode, level+1)  # 그리고 currnode의 child node를 방문
            
    def DFT_preorder(self):
        self.__DFT_preorder_help(self.root)
        
    def __DFT_inorder_help(self, node:TreeNode):
        if currnode == None:
            return
        for i in range(len(currnode.child)):    # idx = 0이면 계속 왼쪽 아래 child 탐색(1)
            if i ==1:                           # child의 child[0]가 None이면 자기 output(2)            
                self.visit(currnode)             # Parent node의 경우 child[0]이 출력되고 나서 출력. 
            self.__DFT_inorder_help(currnode.child[i])
            
    def DFT_inorder(self):
        self.__DFT_inorder_help(self.root)
        
    def __DFT_postorder_help(self, node:TreeNode):
        if currnode == None:
            return
        for i in range(len(currnode.child)):
            self.__DFT_postorder_help(currnode.child[i])
        self.visit(currnode)                     # 맨 왼쪽아래 child부터 탐색. 
```
